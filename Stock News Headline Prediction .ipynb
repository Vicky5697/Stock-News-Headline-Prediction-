```python
#import libraries
import pandas as pd
```

```python
# import data
data = pd.read_csv('Data.csv')
```

```python
#viewing head of data
data.head()
```

```python
# creating train and test based of data
train = data[data['Date']<'20150101']
test = data[data['Date']>'20141231']
```

```python
data1 = train.iloc[:,2:27]  #[rows:,columsns:]
```

```python
data1.replace('[^a-zA-Z]',' ',inplace = True)
```

```python
data1.head()
```

```python
data1.columns
```

```python
new_index = [str(i) for i in range(25)]
new_index
```

```python
data1.columns = new_index
data1.head()
```

```python
data1.index
```

```python
for index in new_index:
    data1[index] = data1[index].str.lower()
data1.head()
    
```

```python
data1.iloc[1,0:25]
```

```python
headlines = []
for i in data1.iloc[1,0:25]:
    headlines.append(i)
    
' '.join(headlines)
```

```python
' '.join([str(i) for i in data1.iloc[1,0:25]])
```

```python
headlines = []

for row in range(0,len(data1)):
    headlines.append(' '.join([str(i) for i in data1.iloc[row,0:25]]))
```

```python
headlines[0:3]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = CountVectorizer(ngram_range=(2,2))
```

```python
traindata_X = cv.fit_transform(headlines)
```

```python
rnd = RandomForestClassifier(n_estimators=200, criterion='entropy')
```

```python
rnd.fit(traindata_X,train['Label'])
```

```python
test.head()
```

```python
test_transform = []

for row in range(0,len(test)):
    test_transform.append(' '.join([str(i) for i in test.iloc[row,2:27]]))
```

```python
test_data = cv.transform(test_transform)
```

```python
y_pred = rnd.predict(test_data)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#Confusion Matrix
cm = confusion_matrix(test['Label'],y_pred)
cm
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# to visualize the confusion matrix 
plt.imshow(cm,cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar() #Creates a colorbar near the figure

labels= ['Positive','Negative']

tick_marks = np.arange(len(labels))
plt.xticks(tick_marks,labels)
plt.yticks(tick_marks,labels)


plt.tight_layout()
plt.xlabel('True Labels')
plt.ylabel('Predicted Label')
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
accuracy_score(test['Label'],y_pred)
```

```python
print(classification_report(test['Label'],y_pred))
```

```python
def plot_confusion_matrix(cm,title='Confusion_matrix'):
    plt.imshow(cm,cmap=plt.cm.Blues)
    plt.title(title)
    plt.colorbar() #Creates a colorbar near the figure

    labels= ['Positive','Negative']

    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks,labels)
    plt.yticks(tick_marks,labels)


    plt.tight_layout()
    plt.xlabel('True Labels')
    plt.ylabel('Predicted Label')
```

```python
plot_confusion_matrix(cm2)
```
